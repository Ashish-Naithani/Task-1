```python
df1=pd.read_excel('Correct.xlsx',index_col=0)
```

```python
#Checking datatypes of columns
df1.info()
```

```python
#Checking amount of null values
df1.isnull().sum()
```

```python
#Editing the values of Income column to create consistency
df1['Income']=df1['Income'].replace('fifty thousand',50000.0)
mean=df1['Income'].describe()[1]
print('Mean of data=',mean)
warnings.filterwarnings('ignore')
```

```python
#Clearing Null values from all columns of the DataFrame
df1=df1.apply(lambda x: x.fillna(round(x.mean(),0)) if x.dtype=='int' or x.dtype=='float' 
            else x.fillna(x.value_counts().index[0]))
```

```python
#Rechecking Null values
df1.isnull().sum()
```

```python
#Checking the inconsistency of format in Dt_customer column
df1['Dt_Customer'].head(10)
```

```python
# Making the formats constant throughout the dataset
df1['Dt_Customer']=df1['Dt_Customer'].astype('datetime64[ns]')
df1['Income']=df1['Income'].astype('float64')
```

```python
#Rechecking datatypes if they are correct for their columns
df1.info()
```

```python
#Checking the amount of duplicated rows presnt in our database
df1['ID'].duplicated().sum()
```

```python
df1.drop_duplicates(inplace=True,keep='first')
```
